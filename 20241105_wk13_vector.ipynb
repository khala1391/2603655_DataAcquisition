### create vectorDB

```python
import chromadb

# Create a client
client = chromadb.Client()

# Create a collection
collection = client.create_collection("movies")

# Add some movie data
collection.add(
    documents=[
        "The Shawshank Redemption is a classic prison drama",
        "Inception is a mind-bending sci-fi thriller",
        "The Godfather is an iconic mafia film",
        "Jurassic Park features dinosaurs in a theme park"
    ],
    metadatas=[
        {"genre": "Drama", "year": 1994},
        {"genre": "Sci-Fi", "year": 2010},
        {"genre": "Crime", "year": 1972},
        {"genre": "Adventure", "year": 1993}
    ],
    ids=["1", "2", "3", "4"]
)

# Query the collection
results = collection.query(
    query_texts=["A movie about dreams"],
    n_results=2
)

# Print results
print("Query Results:")
for i, doc in enumerate(results['documents'][0]):
    print(f"{i+1}. {doc}")
    print(f"   Metadata: {results['metadatas'][0][i]}")
    print()

# Search by metadata
metadata_results = collection.query(
    query_texts=[""],
    where={"genre": "Drama"},
    n_results=1
)

print("Metadata Search Result:")
print(metadata_results['documents'][0][0])
print(f"Metadata: {metadata_results['metadatas'][0][0]}")
```

### recommendation system

```python
import chromadb
from chromadb.utils import embedding_functions
from sentence_transformers import SentenceTransformer

df = pd.read_json('data/aansnook_books.json',orient='table')

client = chromadb.Client()

try:
  client.delete_collection('aansnook_books')
except:
  pass

collection = client.create_collection('aansnook_books')
model = SentenceTransformer('paraphrase-multilingual-MiniLM-L12-v2')

#------------------------------------------
def create_book_text(row):
  return f"{row['title']} {row['author']} {row['description']} {' '.join(row['keywords'])}"

for index,row in df.iterrows():
  book_text = create_book_text(row)
  embedding = model.encode(book_text).tolist()

  collection.add(
      ids = [str(row['id'])],
      embeddings = [embedding],
      metadatas = [{'title':row['title'],
                    'author':row['author'],
                    'price:':row['price']
                    }]
      )

print('books added')

# -----------------

def get_recommendation(book_id,n=5):
  query_book = collection.get(ids=[book_id],include=['embeddings'])
  query_embedding = query_book['embeddings'][0]
  results = collection.query(
      query_embeddings = [query_embedding],
      n_results = n+1
    )
  recommend_ids = [id for id in results['ids'][0] if id != book_id]
  return recommend_ids[:n]

sample_book_id = str(df.iloc[0]['id'])
recommendations = get_recommendation(sample_book_id)
print(f'recommend books for id {sample_book_id}')
print(df.iloc[0]['title'])
print('------------')
for rec_id in recommendations:
  book = df[df['id']==int(rec_id)].iloc[0]
  print(f"- {book['title']} by {book['author']}")

# ------------------------

def semantic_search(query,n=3):
  query_embedding = model.encode(query).tolist()
  results = collection.query(
      query_embeddings=[query_embedding],
      n_results = n
      )
  return results['ids'][0],results['metadatas'][0]

search_query = 'หนังสือเกี่ยวกับการพัฒนาตนเองและความสำเร็จ'
search_results,search_metadatas = semantic_search(search_query)
print('=====-----=====----=====----====')
print(f'ค้นหาหนังสือ: {search_query}')
for id,metadata in zip(search_results,search_metadatas):
  print(f"- {metadata['title']} by {metadata['author']}")

search_query = 'หนังสือเกี่ยวกับการท่องเที่ยว'
# search_query = 'หนังสือเกี่ยวหมูเด้ง'
search_results,search_metadatas = semantic_search(search_query)
print('=====-----=====----=====----====')
print(f'ค้นหาหนังสือ: {search_query}')
for id,metadata in zip(search_results,search_metadatas):
  print(f"- {metadata['title']} by {metadata['author']}")
```

### Predict Rating

```python
import chromadb
from chromadb.utils import embedding_functions
from sentence_transformers import SentenceTransformer

df = pd.read_json('data/aansnook_books.json',orient='table')

client = chromadb.Client()

try:
  client.delete_collection('aansnook_books')
except:
  pass

collection = client.create_collection('aansnook_books')
model = SentenceTransformer('paraphrase-multilingual-MiniLM-L12-v2')

#------------------------------------------

review_count = 0
for index,row in df.iterrows():
  for i,review in enumerate(row['reviews']):
    review_text = review['comment']
    review_id = str(row['id']) + '_' + str(i)
    embedding = model.encode(review_text).tolist()
    collection.add(
        ids= [review_id],
        documents = [review_text],
        embeddings = [embedding],
        metadatas = [{'user':review['user'],
                      'rating':review['rating'
                    ]}]
        )
    review_count += 1
print(f'{review_count} reviews added')

# -------------------------------------------
def search_review(query,n=5):
  query_embedding = model.encode(query).tolist()
  results = collection.query(
            query_embeddings=[query_embedding],
            n_results = n
      )
  return results['metadatas'][0]

search_query = 'หนังสือเล่มนี้ดีที่สุดเท่าที่เคยอ่านมา ผู้เขียนรู้จริง'
search_results = search_review(search_query)
df_search_results = pd.DataFrame(search_results)
print(f"expected rating: {df_search_results['rating'].mean():0.2f}")
```
